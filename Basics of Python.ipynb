### Project Calculator

```python
num1 = int(input("Enter 1st number: "))
num2 = int(input("Enter 2st number: "))
operator = input("Enter operator(+,-,*,/,%): ")

if operator == "+":
    print(f"{num1}{operator}{num2} = {num1+num2}")
elif operator == "-":
    print(f"{num1}{operator}{num2} = {num1-num2}")
elif operator == "*":
    print(f"{num1}{operator}{num2} = {num1*num2}")
elif operator == "/":
    print(f"{num1}{operator}{num2} = {num1/num2}")
elif operator == "%":
    print(f"{num1}{operator}{num2} = {num1%num2}")
else:
    print("Incorrect Operator")
```
