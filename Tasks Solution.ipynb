```python
# Task 13

sentence = input("Write some :").lower()
final_sentence = ""
i=0

for ch in sentence:
    if i%2 == 0:
        final_sentence += ch.upper()
    else:
        final_sentence += ch
    i += 1

print(final_sentence)
```
